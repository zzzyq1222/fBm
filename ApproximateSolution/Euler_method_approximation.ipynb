```python
import sys
sys.path.append('..')
import utils
import EulerMethod as EM
import matplotlib.pyplot as plt
import numpy as np
```

```python
# use Euler-Maruyama method to approximate Ornstein-Uhlenbeck
theta = 0.7
mu = 1.5
sigma = 0.06
x0 = 0
```

```python
timespan = 10
interval = 0.001
n = 10  # number of paths
```

```python
em = EM.EulerMethod()
# Ornstein-Uhlenbeck
ou_process = em.simulateOUProcess(theta, mu, sigma, x0, timespan, interval, n)
utils.draw_n_paths(n, timespan, interval, ou_process)
```

```python
# GBM
mu = 0.5
sigma = 0.5
S0 = 10
```

```python
# Euler-Maruyama method
gbm = em.simulateGBM(mu, sigma, S0)
# utils.draw_n_paths(n, timespan, interval, gbm, 'time(t)', 'y')
fig, ax = plt.subplots()
x = np.linspace(0, timespan, int(timespan / interval))
for i in range(n):
    y = gbm[i]
    ax.plot(x, y, linewidth=0.5)
plt.xlabel('time(t)')
plt.ylabel('$S_t$')
plt.title('Geometric Brownian Motion')
# plt.savefig('../figs/GBM.pdf')
plt.show()
```

```python
# refined Euler method
gbm_refined = em.simulateGBMRefined(mu, sigma, S0)
utils.draw_n_paths(n, timespan, interval, gbm_refined, 'time(t)', 'y')
```

```python
# Bm with drift
mu = 1
sigma = 1
X0 = 0
bmd = em.simulateBMDrift(mu, sigma, X0)
# utils.draw_n_paths(10, 10, 0.001, results, 'time(t)', 'y')
fig, ax = plt.subplots()
x = np.linspace(0, timespan, int(timespan / interval))
for i in range(n):
    y = bmd[i]
    ax.plot(x, y, linewidth=0.5)
plt.xlabel('time(t)')
plt.ylabel('$X_t$')
plt.title('Bm with drift and diffusion')
# plt.savefig('../figs/BMDrift.pdf')
plt.show()
```
